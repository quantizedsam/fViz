# UI Figure Example

> Usage of `fViz.ui.Figure.Figure` class.

## Dependencies

```python
from fViz.ui import Figure
import numpy as np
```

## Plot Figure

```python
# initialize data
xs = np.around(np.linspace(0, 5, 501), 2)
ys = xs**2

# plot parameters
plot_params = {
    'title': 'UI Figure Example',
    'x_label': 'X-axis',
    'y_label': 'Y-axis'
}

# initialize figure
f = Figure(plot_params, xs.tolist())

# plot figure
f.update(xs, ys)
```
